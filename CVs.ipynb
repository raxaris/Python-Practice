```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Output of first 20 rows of data frame

```python
df = pd.read_csv("CV.csv")
```

```python
df.head(20)
```

# Sorting datagrame  by the salary, age and title

```python
df = df.sort_values(by=['Salary', 'Age', 'Title'], ascending=[False, True, True])
df
```

# Calculating minimum, maximum and mean average of the salary of all

```python
salary_min = df['Salary'].min()
salary_max = df['Salary'].max()
salary_mean = df['Salary'].mean()

print("Minimum salary: ${:,.2f}".format(salary_min))
print("Maximum salary: ₸{:,.2f}".format(salary_max))
print("Mean salary: {:,.2f}".format(salary_mean)) 
```

#  Calculating minimum, maximum and mean average of the salary of each sex, specialization and citizenship

```python
grouped_df = df.groupby(['Sex', 'Specialization'])
result = grouped_df.agg({'Salary': ['min', 'max', 'mean']})

result
```

# Graphs

## Chart of the average mean of expected salary each country’s citizenship or of each sex

```python
grouped_df = df.groupby(['Sex'])['Salary'].mean()

grouped_df.plot(kind='bar')

plt.title('Mean Average of Expected Salary by Sex')
plt.xlabel('Sex')
plt.ylabel('Mean Expected Salary')

plt.show()
```

## Chart of the minimum, maximum, and mean average of the salary of all


```python
min_salary = df['Salary'].min()
max_salary = df['Salary'].max()
mean_salary = df['Salary'].mean()

plt.bar(['Minimum', 'Maximum', 'Mean Average'], [min_salary, max_salary, mean_salary])

plt.title('Minimum, Maximum, and Mean Average of Salary')
plt.xlabel('Statistics')
plt.ylabel('Salary')

plt.show()
```

##  Chart of minimum, maximum and mean average of the salary of all each specialization 

```python
grouped_df = df.groupby(['Specialization'])['Salary'].agg(['min', 'max', 'mean'])

grouped_df.plot(kind='bar')

plt.title('Minimum, Maximum, and Mean Average of Salary by Specialization')
plt.xlabel('Specialization')
plt.ylabel('Salary')

plt.show()
```

## Chart of minimum, maximum and mean average of ages


```python
min_age = df['Age'].min()
max_age = df['Age'].max()
mean_age = df['Age'].mean()


plt.boxplot(df['Age'])

plt.plot([1, 1], [min_age, min_age], 'go', markersize=10)
plt.plot([1, 1], [max_age, max_age], 'ro', markersize=10)
plt.plot([1, 1], [mean_age, mean_age], 'bo', markersize=10)

plt.title('Minimum, Maximum, and Mean Average of Age')
plt.xticks([1], ['Age'])
plt.ylabel('Age')

plt.show()
```

## Graph of minimum, maximum and mean average title

```python
salary_stats = df.groupby('Title')['Salary'].agg(['min', 'max', 'mean'])

title_counts = df['Title'].value_counts()

fig, ax = plt.subplots()
ax.bar(title_counts.index, salary_stats['mean'], yerr=[
       salary_stats['mean'] - salary_stats['min'], salary_stats['max'] - salary_stats['mean']])
ax.set_xticks(range(len(title_counts.index)))
ax.set_xticklabels(title_counts.index, rotation=90)
ax.set_ylabel('Salary')
ax.set_title('Minimum, Maximum, and Mean Average of Salary for Each Title')

plt.show()
```
